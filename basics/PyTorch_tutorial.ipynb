# Tensors 

Tensors are a specialized data structure that are very similar to arrays and matrices. In PyTorch, we use tensors to encode the inputs and outputs of a model, as well as the model's paramenters.

Tensors are similar to NumPy's ndarrays, except that tensors can run on GPUs or other hardware accelerators. In fact, tensors and NumPy arrays cna often share the same underlying memory, eleiminating the need to copy data. Tensor are also optmized for automatic differentiation.

## Initializing a Tensor

```python
data = [[1,2], 
       [3, 4]]
x_data = torch.tensor(data)
```

```python
x_data
```

Tensors can be created from NumPy arrays(and vice versa)

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

## Atributes of a Tensor

Tensor attributes describe their shape, datatype, and the device on which they are stored.

```python
tensor = torch.rand(3,4)
print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

## Operations on Tensors



Over 100 tensor operations, including arithmetic, linear algebra, matrix manipulation(transposing, indexing, slicing), sampling and more...

### NOTE:

Each of these operations can be run on the GPU. By default, tensors are created on the CPU. We need to explicitly move tensor to the GPU using .to method. Keep in mind that copying large tensors across devices can be expensive in terms of time and memory!

```python
# We move our tensor to the GPU if available
if torch.cuda.is_available():
    tensor = tensor.to("cuda")
```

Examples of operations on tensor:

```python
print(f"Device tensor is stored on: {tensor.device}")
```

```python
tensor = torch.ones(4,3)
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
tensor[:,1] = 0
print(tensor)
```
